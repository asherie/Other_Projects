### Analysis

Datafile has five columns: year (1000 to 1998) and intensity ("forcing") contributions from volcanic, solar, greenhouse and aerosols.
Positive values mean net energy input (warming); negative values mean net energy output (cooling). Units are W/m^2. 

Origin of data:  
(1) volcanic—conductivity and sulfate records from ice core samples from Greenland and Antarctica together with volcanism records.  
(2) solar—abundance of cosmogenic isotopes (mainly beryllium-10 supplemented with carbon-14 data). Seems to correlate with sunspot activity (see <a href='#Appendix-A'>Appendix A</a>)  
(3) greenhouse—abundance of carbon dioxide and othet trace greenhouse gases together with a "standard equivalent radiative forcing model". It is not clear which model is used (this one? https://en.wikipedia.org/wiki/Radiative_forcing#Example_calculations) nor where the CO2 and other gas concentrations are taken from.  
(4) aerosol—tropospheric aerosols. Presumably, volcanic aerosols are not included, although it is not clear how volcanic aerosols are separated from other types of aerosols. One way to distinguish between the two categories of aerosols is to assume that volcanic aerosols are predominantly in the stratosphere (http://volcanoes.usgs.gov/hazards/gas/s02aerosols.php), while all other aerosols are in the troposphere, but this is a simplistic assumption given our lack of knowledge about natural aerosols. (http://www.carbonbrief.org/blog/2013/11/natural-aerosols-complicate-climate-understanding/)

### Comments


(1) Volcanic activity provides large negative forcing after eruptions. This makes sense as the sulfur particles produced by the eruptions cool the atmosphere by scattering sunlight. The largest cooling effect is around 1250, although it is unclear which volcano is responsible. (See http://www.wired.com/2012/02/the-mysterious-missing-eruption-of-1258-a-d/)
Also, the net cooling effect of sulfur-based particles produced by volanic eruptions is the inspiration for geoengineering solutions to offset rising temperatures i.e., spray the upper atmosphere with sulfuric acid. (See http://www.technologyreview.com/featuredstory/511016/a-cheap-and-easy-plan-to-stop-global-warming/)

(2) Solar forcing is small and averages about zero. There appears to be cyclic variations in the forcing, which correlates with sunspot activity (see <a href='#Appendix-A'>Appendix A</a>). 

(3) Greenhouse forcing is very small (smaller than the solar forcing) and fluctuates about zero until about 1800. After that it rapidly climbs upwards (i.e., contributes to warming) until the data ends in 1998. Note that the industrial revolution, which resulted in large-scale production of greenhouse gases, began in the late 1700's.

(4) Aerosol data is zero until 1890, when it decreases rapidy until 1998.

There appears to be a correlation between the solar forcing and the sunspot number, but it is obscured by the 11-year solar cycle. Therefore, smooth the sunspot data before calculating the correlation with the solar forcing.

Take a simple moving average of the number of sunspots. Choose 11 years as moving average window.

```python
def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')

interval_value = 11
sp_smooth = movingaverage(sunspot['Sunspots'], interval_value) #11 year moving average

plt.plot(sunspot['Year'], sunspot['Sunspots'], label = 'Raw')
plt.plot(sunspot['Year'], sp_smooth, label='%i-year moving average' %interval_value )
plt.title('Smoothing of Sunspot Data')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize='x-large')
plt.xlim(1700,2014)
```

Results look reasonable. Now plot moving average with solar forcing.

```python
fig, ax1 = plt.subplots()
ax1.plot(forcing['Year'], forcing['Solar'], 'y-', linewidth=2.0)
ax1.set_xlabel('Year', size=14)
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Solar Forcing ($W m^{-2}$)', color='y', size=14)
for tl in ax1.get_yticklabels():
    tl.set_color('y')


ax2 = ax1.twinx()
ax2.plot(sunspot['Year'], sp_smooth, 'b-')
ax2.set_ylabel('Moving avg of Sunspot #', color='b', size=14)
for tl in ax2.get_yticklabels():
    tl.set_color('b')
# Use x-axis range for which there is both solar forcing and sunspot number data
plt.xlim(1700,1998)    
plt.show()
```

Correlation is now more evident. To calculate the cross-correlation, select subset of solar forcing and moving average data for same years (1700-1998).

```python
log_ga_net_select=log_ga_net[800:999]
print log_ga_net_select # check values are sensible
```

```python
x=year_select
y=log_ga_net_select
p, cov = np.polyfit(x, y, 1, cov=True)  # coefficients and covariance matrix
yfit = np.polyval(p, x)                 # evaluate the polynomial at year_select
perr = np.sqrt(np.diag(cov))       # standard-deviation estimates for each coefficient
R2 = np.corrcoef(x,y)[0, 1]**2    # coefficient of determination between x and y
resid = y - yfit

# if yerr, standard deviation of observation is known, can use this:
# chi2red = np.sum((resid/yerr)**2/(y.size - 2) #reduced chisquared (measure of degrees of freedom)
# should be close to 1 or else something is off

# Since yfit <0, the other formulation of chi-squared np.sum((resid**2)/yfit)/(y.size - 2) does not work.

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'co')
plt.plot(x, yfit, 'r', linewidth=3, color=[1, 0, 0, .5])
plt.xlabel('Year', fontsize=16)
plt.ylabel('log(Anthropogenic Contribution)', fontsize=16)
plt.title('$y = %.2f \pm %.2f + (%.2f \pm %.2f)x \; [R^2=%.2f]$'
         %(p[1], perr[1], p[0], perr[0], R2), fontsize=20, color=[0, 0, 0]) 
# If reduced chi-squared is known, use version below
# plt.title('$y = %.2f \pm %.2f + (%.2f \pm %.2f)x \; [R^2=%.2f,\, \chi^2_{red}=%.1f]$'
#          %(p[1], perr[1], p[0], perr[0], R2, chi2red), fontsize=20, color=[0, 0, 0]) 
plt.xlim((1800, 2000))
plt.show()
```

Fit is decent, but there is clearly a cyclic component to the increase after 1900. Also, from 1800 to 1900 the data increases stepwise, which is probably a reflection of the resolution available when looking into the past.  

Since the errors in the data are not known, a simple reduced chi-squared test is not possible. A deeper issue is that there are probably correlations in the data, which would invalidate the use of a chi-squared test.
