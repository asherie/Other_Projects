### Introduction

Datafile has five columns: time (in reverse chronological order; from 2011 to 2009); latitude (degrees); longitude (degrees);  depth (in km); and magnitude. The moment magnitude is a dimensionless number, which is calculated from the seismic moment (See https://en.wikipedia.org/wiki/Moment_magnitude_scale); roughly, it is a logarithm measure of the energy of the earthquake. Its definition differs slightly from the Richter scale, which is no longer used.

2. Depth is shallow (less than 30 km). Makes sense as most earthquakes are shallow (see https://en.wikipedia.org/wiki/Earthquake).   
3. Magnitude is below 5 with most below 3. Again, makes sense. If there had been a large earthquake in 2009-2011 near New York, I would have remembered it as I've lived in New York since 2004.

Depth increases with magnitude, but it also seems that the deepest quakes have an intermediate magnitude. Therefore, try a polynomial (cubic) fit and LOESS. 

```python
# Adapted from Practical 1
best_fit4 = np.polyfit(quake_mag['mag'],quake_mag['depth'],3)

def loess( x, h, xp, yp ):
    w = np.exp( -0.5*( ((x-xp)/h)**2 )/np.sqrt(2*np.pi*h**2) )

    b = sum(w*xp)*sum(w*yp) - sum(w)*sum(w*xp*yp)
    b /= sum(w*xp)**2 - sum(w)*sum(w*xp**2)
    a = ( sum(w*yp) - b*sum(w*xp) )/sum(w)

    return a + b*x

s1, s2 = [], []
for k in quake_mag['mag']:
    s1.append( loess( k,  1, quake_mag['mag'], quake_mag['depth'] ) )
    s2.append( loess( k, 100, quake_mag['mag'], quake_mag['depth'] ) )

plt.plot(quake_mag['mag'], quake_mag['depth'],'o',color="white")
fit1, = plt.plot(quake_mag['mag'],np.polyval(best_fit4,quake_mag['mag']),'r-', label='cubic')
fit2, = plt.plot(quake_mag['mag'], s1, label='loess1')
fit3, = plt.plot(quake_mag['mag'], s2, label='loess2')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize='x-large')
plt.xlabel( "Magnitude", fontsize=14 )
plt.ylabel( "Depth (km)", fontsize=14)
plt.title( "New York Area Earthquakes", fontsize=16 )
plt.ylim(-4,30)
plt.xlim(0.5,4.5)
```

LOESS fits give a monotonic increase of depth vs. magnitude. The intermediate maximum, which is captured by the cubic fit, may just be noise.

### Answer 4

Use distance function provided by Aaron Golden. Modified to use numpy instead of math, so it works better with lists (minimze type errors).

```python
def distance_on_sphere(lat1, long1, lat2, long2):

    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = np.pi/180.0
        
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
        
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
        
    # Compute spherical distance from spherical coordinates.
        
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
    
    cos = (np.sin(phi1)*np.sin(phi2)*np.cos(theta1 - theta2) + 
           np.cos(phi1)*np.cos(phi2))
    arc = np.arccos( cos )

    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length
    # radius in km = 6373
    return arc*6373
```

```python
# Test that dist works
dist_NYC_BOS=distance_on_sphere(40.7,-74.0,42.35,-71.1)
dist_BOS_NYC=distance_on_sphere(42.35,-71.1,40.7,-74.0)
print dist_NYC_BOS
print dist_BOS_NYC
# Distance in km; it is about 340 km by car. A little smaller value than this is what I'd expect.
# Also, order of cities doesn't matter
```

Perform K-clustering analysis.

```python
# Kmeans only works with float values, so first remove time from dataframe

Z_notime = quake.drop('time',1) 

# Kmeans takes an array, so covert reduced dataframe to array
X = pd.np.array(Z_notime)

# print X; check that it looks reasonable. I also looked at the header and file looked good.
from sklearn.cluster import KMeans  # load in Kmeans
n_clusters = 4 # number of clusters
k_means = KMeans(n_clusters, random_state=0) # initializing kmeans object, 3 cluster centroids, random start points.
k_means.fit(X)   # perform K-means against entire data matrix
y_pred = k_means.predict(X)  # predict cluster associations
plt.scatter(X[:, 1], X[:, 0], c=y_pred, s=80) 
# Use same axis as for unclustered plot
plt.xlim(-79,-70)
plt.ylim(37,46)
plt.title('K-means clustering, k = %i' % (n_clusters) );
```

Try clustering with DBSCAN. Perhaps it will do a better job than K-clustering.

```python
from sklearn.cluster import DBSCAN   # load in DBSCAN
from sklearn import metrics          # load in the very useful methods to determine how effective a method has been

# recall X is from quake with time dropped
# here eps = the distance constraint, min_samples = the neighborhood count threshold
# so we create a data object db that represents a DBSCAN analysis on the data 

db = DBSCAN(eps=1.25, min_samples=5).fit(X)

labels = db.labels_   # determine the class labels for each sample (row) of the matrix
n_clusters_= len(set(labels)) - (1 if -1 in labels else 0)  # the total number of clusters found...

print n_clusters_
print labels
```

Four clusters found, but it looks like many points are not in a cluster (that is what -1 means in labels). 

```python
print("Silhouette Coefficient: %0.3f"% metrics.silhouette_score(X, labels))
```

A terrible value. I varied the parameters eps and min_samples, but the coefficient never got above 0.5 (and the large values correspond to about 20 clusters i.e., no clustering). Seems to me that the quake data does not cluster well in 4-d space. The plot shows, however, that Ramapo line is associated with some quakes.

Plot the data using Aaron's algorithm from Practical 7. Add the Ramapo line too.

```python
unique_labels = set(labels)
plt.figure(figsize=(8, 8))
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)   #   creates 'truth' array for label associations
core_samples_mask[db.core_sample_indices_] = True   #  fills 'truth' array with actual cluster associations
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 1], xy[:, 0], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 1], xy[:, 0], 'o', markerfacecolor=col,
            markeredgecolor='k', markersize=6)
    
plt.xlim(-79,-70)
plt.ylim(37,46)
plt.xlabel('Longitude (degrees)', fontsize=14)
plt.ylabel('Latitude (degrees)', fontsize=14)
plt.title('Estimated number of clusters: %d' % n_clusters_, fontsize=16)
plt.plot(ramp_line['longitude'], ramp_line['latitude'], c='blue') 
plt.show()
```

Two of the clusters lie near the Ramapo fault line, but I don't think that the evidence for clustering is strong. As I mentioned above, the association between the fault line some quakes can be seen by eye.
